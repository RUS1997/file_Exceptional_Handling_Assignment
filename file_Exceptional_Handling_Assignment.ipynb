#Q1. What is the difference between interpreted and compiled languages?
  - Compiled languages translate the entire source code into machine code before execution, resulting in faster runtime performance.
  - Interpreted languages, like Python, execute code line-by-line, which can be slower but offers more flexibility.
#Q2. What is exception handling in Python?
  - Exception handling in Python involves managing errors that occur during program execution. It uses try, except, else, and finally blocks to catch and handle exceptions gracefully, preventing program crashes.
#Q3. What is the purpose of the finally block in exception handling?
  - The finally block contains code that executes regardless of whether an exception occurred or not. It's typically used for cleanup actions, like closing files or releasing resources.
#Q4. What is logging in Python?
  - Logging in Python is a way to track events that happen during program execution. The built-in logging module allows developers to record messages with different severity levels, which can be invaluable for debugging and monitoring applications.
#Q5. What is the significance of the_del__ method in Python?
  - The __del__ method is a special method called a destructor. It's invoked when an object is about to be destroyed, allowing for cleanup actions like releasing external resources.

  - import module: Imports the entire module, and you access its functions or classes with the module name prefix (e.g., module.function())

#Q9. What is the difference between multithreading and multiprocessing?
  - Multithreading involves multiple threads within a single process sharing the same memory space. It's suitable for I/O-bound tasks but can be limited by the Global Interpreter Lock (GIL) in Python.
  - Multiprocessing involves multiple processes, each with its own memory space. It's ideal for CPU-bound tasks and can bypass the GIL, allowing true parallelism.

#Q11. What is memory management in Python?

#Q13. Why is memory management important in Python?

#Q14. What is the role of try and except in exception handling?

#Q15. How does Python's garbage collection system work?
  - Python employs automatic memory management through a combination of reference counting and a cyclic garbage collector.
  - Reference Counting: Each object keeps track of the number of references to it. When this count drops to zero, the memory occupied by the object is deallocated.
  - Cyclic Garbage Collector: To handle reference cycles (where objects reference each other), Python's gc module periodically identifies and collects these unreachable objects.
#Q16. What is the purpose of the else block in exception handling?
  - The else block in Python's exception handling is executed only if the try block does not raise an exception. It's useful for code that should run only when no errors occur, keeping the try block concise and avoiding the accidental catching of exceptions that weren't raised by the code being protected.
#Q17. What are the common logging levels in Python?
  - Python's logging module defines several standard logging levels, each associated with a numeric value indicating the severity of the event:
  - DEBUG (10): Detailed information, typically of interest only when diagnosing problems.
  - INFO (20): Confirmation that things are working as expected.
  - WARNING (30): An indication that something unexpected happened, or indicative of some problem in the near future. The software is still working as expected.
  - ERROR (40): Due to a more serious problem, the software has not been able to perform some function.
  - CRITICAL (50): A serious error, indicating that the program itself may be unable to continue running.
#Q18. What is the difference between os.fork() and multiprocessing in Python?
 - os.fork(): This function creates a new child process by duplicating the current process. It's available only on Unix-like systems and is a low-level method that doesn't provide additional abstractions.
 - multiprocessing module: This is a high-level module that allows the creation of processes using an API similar to the threading module. It provides better portability (works on Windows and Unix), easier process management, and avoids issues related to the Global Interpreter Lock (GIL) by using separate memory spaces for each process.
 #Q19. What is the importance of closing a file in Python?
 - Resource Management: It frees up system resources associated with the file.
 - Data Integrity: Ensures that all buffered data is written to the disk, preventing data loss or corruption.
 - Avoiding File Corruption: Properly closing files helps prevent file corruption, especially in write operations.
 #Q20. What is the difference between file.read() and file.readline() in Python?
 - file.read(): Reads the entire content of the file into a single string. It's useful when you need to process the whole file at once.
 - file.readline(): Reads one line from the file at a time, returning it as a string. This is beneficial when dealing with large files, allowing you to process one line at a time without loading the entire file into memory.

#Q23. What are the challenges associated with memory management in Python?

  - Memory Leaks: Long-lived objects or global variables can prevent memory from being freed.
  - Performance Overhead: The garbage collector can introduce performance overhead, especially in programs with a large number of objects or complex object graphs.

#Q25. Why is it important to use multithreading in certain applications?
  - Multithreading is beneficial in applications where tasks are I/O-bound, such as reading from or writing to files, network operations, or user interface updates. It allows multiple threads to run concurrently, improving the responsiveness and efficiency of the application.
```python
#Q18. How would you implement a basic logging setup that logs to a file with rotation after IMB?
import logging
from logging.handlers import RotatingFileHandler

logger = logging.getLogger("RotatingLog")
logger.setLevel(logging.INFO)

handler = RotatingFileHandler("rotating.log", maxBytes=1024*1024, backupCount=2)
logger.addHandler(handler)

for i in range(1000):
    logger.info("Logging line %d", i)
```

```python
#Q21. Write a Python program that reads a file and prints the number of occurrences of a specific word.
word_to_count = "hello"
count = 0

with open("sample.txt", "r") as f:
    for line in f:
        count += line.lower().split().count(word_to_count)

print(f"'{word_to_count}' occurred {count} times.")
```
